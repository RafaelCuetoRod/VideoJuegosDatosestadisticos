# **Histograma de intensidad de colores de la imagen (Azul, Verde, Rojo)**

```python
plt.figure(figsize=(8,6))
plt.hist(b.ravel(),bins=256,range=(0,256),color='b',label='blue')
plt.hist(g.ravel(),bins=256,range=(0,256),color='g',label='green')
plt.hist(r.ravel(),bins=256,range=(0,256),color='r',label='red')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(b.ravel(),bins=256,range=(0,256),color='b',label='blue',alpha=0.5)
plt.hist(g.ravel(),bins=256,range=(0,256),color='g',label='green',alpha=0.5)
plt.hist(r.ravel(),bins=256,range=(0,256),color='r',label='red',alpha=0.5)
plt.legend()
plt.show()
```

La distribuci칩n de intensidad de color azul es unimodal sesgada hacia la derecha. La mayoria de las intencidades esta asociada a 191 (media si tomamos el grado de intensidad entre 130 y 250, y se logra visualisar que la moda es 196), a demas el 50% de los pixceles (313920) son menores a 178. sin embargo considerando la desviacion estandar (60) las intensidades m치s probables estan en el intervalo (131,251)
